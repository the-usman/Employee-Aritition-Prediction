```python
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
def multivarinent(data, x_column, y_column, color='Attrition', graphType='Histogram'):
    if graphType == 'Histogram':
        fig = px.histogram(data, x=x_column,y=y_column, color=color)
        fig.show()
    elif graphType == 'Scatter':
        fig = px.scatter(data, x=x_column, y=y_column, color=color)
        fig.show()
    elif graphType == 'Violin':
        fig = px.violin(data, x=x_column, y=y_column, color=color)
        fig.show()
    elif graphType == 'Density':
        plt.figure(figsize=(14, 8))
        sns.kdeplot(data[x_column], label=x_column)
        sns.kdeplot(data[y_column], label=y_column)  
        plt.legend()
        plt.show()
    elif graphType == 'Countplot':
        plt.figure(figsize=(14, 8))
        plt.subplot(121)
        sns.countplot(data=data, x=x_column, hue=color)
        plt.subplot(122)
        sns.countplot(data=data, x=y_column, hue=color)  # Changed x_column to y_column
        plt.show()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class FeatureEngineering (BaseEstimator, TransformerMixin) :
    def fit(self, X, y=None) :
        return self

    def transform(self, X, y=None) :
        bin_edages = [-float('inf'), 10, 20, float('inf')]
        bin_labels = ['close', 'medium', 'far']
        X['DistanceHome'] = pd.cut(X['DistanceFromHome'], bins=bin_edages, labels=bin_labels)
        X['DiffMonthlyRateMonthlyIncome'] = X['MonthlyRate'] - X['MonthlyIncome']
        X['TimeInInductry'] = X['TrainingTimesLastYear'] + X['TotalWorkingYears']
        return X 
```

```python
from sklearn.model_selection import train_test_split

inputs = data.drop(['Attrition'], axis=1)
targets = data['Attrition']

X_train, X_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

trf1 = ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output=False, drop='first'), ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole',
       'MaritalStatus', 'Over18', 'OverTime'])
], remainder='passthrough', verbose_feature_names_out=False).set_output(transform='pandas')
```

```python
scaledata = trf1.fit_transform(X_train)
```

```python
# Standradiztion  
scaledata.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

trf2 = ColumnTransformer([
    ('scaler', MinMaxScaler(), scaledata.columns)
], remainder='passthrough', verbose_feature_names_out=False).set_output(transform='pandas')
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline( trf1, trf2)
```

```python
X_train = pipe.fit_transform(X_train)
X_test = pipe.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import r2_score
```

```python
def SelectModel (X_train, y_train, X_test, y_test) :
    model_dict = {
        "Logistic" : LogisticRegression(),
        "XGB" : XGBClassifier(random_state=42),
        "Light" : LGBMClassifier(random_state=42),
        "Cat" : CatBoostClassifier(random_state=42, verbose=False)
    }
    

    for model_name, model in model_dict.items() :
        model.fit(X_train, y_train)
        print(roc_auc_score(y_test,model.predict(X_test)))
        print(r2_score(y_test,model.predict(X_test)))
```
